```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 999)
```

```python
listings_1604 = pd.read_csv('../data/listings_2016-04-03.csv.gz', compression='gzip')
listings_1605 = pd.read_csv('../data/listings_2016-05-02.csv.gz', compression='gzip')

list_1604 = listings_1604.copy()
list_1605 = listings_1605.copy()

list_1604.head()
```

```python
list_1604["neighbourhood"].unique()
```

```python
list_1604 = list_1604.loc[:, ["id", "listing_url", "name", "latitude", "longitude"]]
list_1605 = list_1605.loc[:, ["id", "listing_url", "name", "latitude", "longitude"]]
```

```python
list_1604 = list_1604.sort_values(by="id")
# list_1604.join(list_1605)
list_1604.head()
```

```python
list_1604.columns = ["id", "url 04", "name 04", "lat 04", "lon 04"]
list_1604.head()
```

```python
joined = list_1604.join(list_1605.set_index("id"), on = "id")
joined.head()
```

```python
joined["same"] = (joined["url 04"] == joined["listing_url"]) & (joined["name 04"] == joined["name"]) & (joined["lat 04"] == joined["latitude"]) & (joined["lon 04"] == joined["longitude"])
joined.head()
```

```python
not_same = joined[joined["same"] == False]
not_same = not_same[~not_same["listing_url"].isna()]
not_same["same location"] = (not_same["lat 04"] == not_same["latitude"]) & (not_same["lon 04"] == not_same["longitude"])
why_are_you_diff = not_same[not_same["same location"] == False]
```

```python
why_are_you_diff["diff"] = why_are_you_diff["lat 04"] - why_are_you_diff["latitude"]
why_are_you_diff.shape
```
